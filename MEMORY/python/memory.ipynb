```python

import random
import pygame
import time


class button():
    def __init__(self,color,x,y,width,height,tsize,text='',found=None):
        self.color=color
        self.x=x
        self.y=y
        self.width=width
        self.height=height
        self.text=text
        self.tsize=tsize
        self.found=found

    def drawbutton(self,WIN,outline=None):
        if self.found==True:
            self.color=LIGHTBLUE
            self.text=''
        if outline:
            pygame.draw.rect(WIN,outline,(self.x-2,self.y-2,self.width+4,self.height+4),0)
        pygame.draw.rect(WIN,self.color,(self.x,self.y,self.width,self.height),0)
        if self.text != '':
            myfont = pygame.font.SysFont('Comic Sans MS', self.tsize)
            text = myfont.render(str(self.text), 1, (0, 0, 0))
            WIN.blit(text,(self.x+(self.width/2-text.get_width()/2),self.y+(self.height/2-text.get_height()/2)))

    def isOverSquare(self,pos):
        if pos[0]>self.x and pos[0]<self.x+self.width:
            if pos[1]>self.y and pos[1]<self.y+self.height:
                return True
        return False




class memory:
   
    def __init__ (self,row,col,lifes):
        self.pairs=int((row*col)/2)
        self.row=row
        self.col=col
        self.behindBoard=[]
        self.board=[]
        self.lifes=lifes
        self.maxlifes=lifes
        self.createBehindBoard()
        self.createBoard()  
        
    def createBehindBoard(self):
        for i in range(self.pairs):
            self.behindBoard.append([0,i])
            self.behindBoard.append([0,i])
        self.shuffleBehindBoard()
    
    def shuffleBehindBoard(self):
        for i in range(2*self.pairs):
            self.behindBoard[i][0]=random.uniform(0, 1)
        self.behindBoard.sort()
        y=0
        board2=[]
        for i in range(self.row):
            tmp=[]
            for j in range(self.col):
                tmp.append(self.behindBoard[y][1])
                y+=1
            board2.append(tmp)
        self.behindBoard=board2

    def printBehindBoard(self):
        ans=""
        for i in range(len(self.behindBoard)):
            for j in range(len(self.behindBoard[0])):
                ans+=str(self.behindBoard[i][j])
                ans+="|"
            ans+=" row "
            ans+=str(i)
            ans+="\n"
        print(ans)
        
    def createBoard(self):
        
        for i in range(len(self.behindBoard)):
            tmp=[]
            for j in range(len(self.behindBoard[0])):
                tmp.append(' ')
            self.board.append(tmp)
        
    def printBoard(self):
        ans=""
        for i in range(len(self.board)):
            ans+="|"
            for j in range(len(self.board[0])):
                ans+=str(self.board[i][j])
                ans+="|"
            ans+=" row "
            ans+=str(i)
            ans+="\n"
        print(ans)

    def guess(self,r1,c1,r2,c2):
        if (r1,c1)!=(r2,c2):
            if self.behindBoard[r1][c1]==self.behindBoard[r2][c2]:
                self.board[r1][c1]=self.behindBoard[r1][c1]
                self.board[r2][c2]=self.behindBoard[r1][c1]
                return True
            self.lifes-=1
        return False
                           
    def createButtons(self):
        if 0<self.row<=6:
            tsize=50
        elif 7<=self.row<=8:
            tsize=30
        else:
            tsize=20
        buttons=[]
        w=int((WIDTH-10)/self.col)
        h=int((HEIGHT-100)/self.row)
        for i in range(self.row):
            for j in range(self.col):
                x=w*j+10
                y=h*i+100
                buttons.append(button(LIGHTGRAY,x,y,w-10,h-10,tsize,'',False))
        return buttons

    def checkGameOver(self,b):
        count=0
        x=0
        for i in range(self.row):
            for j in range(self.col):
                if b[x].found==True:
                    count+=1
                x+=1
        if count==self.col*self.row:
            return True
        return False

    def checkStarted(self,b):
        if self.lifes<self.maxlifes:
            return True
        count=0
        y=0
        for i in range(self.row):
            for j in range(self.col):
                if b[y].found==True:
                    count+=1
                y+=1
        if count>0:
            return True
        return False

        
pygame.init()
pygame.font.init()
WIDTH,HEIGHT =600,700  
WIN=pygame.display.set_mode((WIDTH,HEIGHT))
pygame.display.set_caption("LIFE")
PINK=(255,0,127)
RED=(255,0,0)
ORANGE=(255,185,0)
YELLOW=(255,255,0)
GREEN=(0,205,0)
CYAN=(0,255,255)
BLUE=(0,0,255)
PURPLE=(120,0,160)
WHITE=(255,255,255)
BLACK=(0,0,0)
LIGHTGRAY=(200,200,200)
LIGHTBLUE=(50,150,255)

def redrawWindow(win,g,count1,count2,buttons=None):
    WIN.fill(LIGHTBLUE)           
    for i in range (len(buttons)):
        buttons[i].drawbutton(WIN)
      
    txt= "LIFES LEFT: "
    txt+=str(g.lifes)
    myfont = pygame.font.SysFont('Comic Sans MS', 20)
    text = myfont.render(txt,1,BLACK )
    WIN.blit(text,(350 +(100-text.get_width()/2),30+((30)-(text.get_height()/2))))

    txt ="GAMES WON: "
    txt+=str(count1)
    txt+=" / "
    txt+=str(count2)
    myfont = pygame.font.SysFont('Comic Sans MS', 20)
    text = myfont.render(txt,1,BLACK )
    WIN.blit(text,(40 +(100-text.get_width()/2),30+((30)-(text.get_height()/2))))
    pygame.display.update()
   

def pickSize(WIN):
    WIN.fill(LIGHTBLUE)
    size=[]
    exp=["3x4","4x6","6x6","6x8","8x8", "10x10"]
    size.append(button(PURPLE,25,50,250,100,30,"SMALL 1"))
    size.append(button(PINK,325,50,250,100,30,"SMALL 2"))
    size.append(button(ORANGE,25,250,250,100,30,"MEDIUM 1"))
    size.append(button(YELLOW,325,250,250,100,30,"MEDIUM 2"))
    size.append(button(GREEN,25,450,250,100,30,"LARGE 1"))
    size.append(button(BLUE,325,450,250,100,30,"LARGE 2"))
    

    for i in range(len(size)):
        pygame.draw.rect(WIN,size[i].color,(size[i].x,size[i].y,size[i].width,size[i].height),0)
        myfont = pygame.font.SysFont('Comic Sans MS', 40)
        text=myfont.render(size[i].text,1,BLACK )
        WIN.blit(text,(size[i].x +(size[i].width/2-text.get_width()/2),size[i].y+(size[i].height/3-text.get_height()/3)))
        myfont = pygame.font.SysFont('Comic Sans MS', 20)
        text = myfont.render(exp[i],1,BLACK )
        WIN.blit(text,(size[i].x +(size[i].width/2-text.get_width()/2),size[i].y+((size[i].height/3)*2-(text.get_height()/3))))
    
    pygame.display.update()
    return size


def gameOver(g,b,txt):
    WIN.fill(LIGHTBLUE)
    for i in range (len(b)):
        b[i].drawbutton(WIN)
    button(WHITE,100,100,400,100,30,txt,False).drawbutton(WIN)
    button(WHITE,100,200,400,200,30,"Do you want to play again?",False).drawbutton(WIN)
    button(WHITE,100,400,400,150,30,'',False).drawbutton(WIN)
    button(GREEN,150,400,100,50,30,"YES",False).drawbutton(WIN)
    button(RED,350,400,100,50,30,"NO",False).drawbutton(WIN)
    b.append(button(GREEN,150,400,100,50,30,"YES",False))
    b.append(button(RED,350,400,100,50,30,"NO",False))
    pygame.display.update()

def main():
    
    run=True
    count1=0
    count2=0
    while run:
        size=False
        r=2
        c=2
        l=2
        while size==False and run==True:
            s=pickSize(WIN)
            for event in pygame.event.get():
                if event.type==pygame.QUIT:
                    size=True
                    run=False
                    pygame.quit() 
                if event.type==pygame.MOUSEBUTTONDOWN:
                    pos=pygame.mouse.get_pos()
                    if s[0].isOverSquare(pos):
                        r=3
                        c=4
                        l=6
                        size=True
                    elif s[1].isOverSquare(pos):
                        r=4
                        c=6
                        l=18
                        size=True
                    elif s[2].isOverSquare(pos):
                        r=6
                        c=6
                        l=30
                        size=True
                    elif s[3].isOverSquare(pos):
                        r=6
                        c=8
                        l=60
                        size=True
                    elif s[4].isOverSquare(pos):
                        r=8
                        c=8
                        l=100
                        size=True
                    elif s[5].isOverSquare(pos):
                        r=10
                        c=10
                        l=140
                        size=True
        restart=False
        m=memory(r,c,l)
        b=m.createButtons()
        b.append(button(WHITE,250,30,100,60,20,"MENU",False))
        while run and restart==False:
            redrawWindow(WIN,m,count1,count2,b)
            clicked=0
            r1=None
            r2=None
            c1=None
            c2=None
            spot1=None
            spot2=None
            while clicked<2 and run and restart==False:
                for event in pygame.event.get():
                    if event.type==pygame.QUIT:
                        run=False
                        pygame.quit()  
                    if event.type==pygame.MOUSEBUTTONDOWN:
                        pos=pygame.mouse.get_pos()
                        x=0
                        for i in range(m.row):
                            for j in range(m.col):
                                if clicked<2:                                    
                                    if b[x].isOverSquare(pos):
                                        if clicked==0:
                                            r1=i
                                            c1=j
                                            spot1=x
                                        elif clicked==1:
                                            r2=i
                                            c2=j
                                            spot2=x
                                        b[x].text=str(m.behindBoard[i][j])
                                        clicked+=1
                                x+=1
                                redrawWindow(WIN,m,count1,count2,b)
                        if b[x].isOverSquare(pos):
                            restart=True
                            start=m.checkStarted(b)
                            if start==True:
                                count2+=1
                if clicked==2 and restart==False:
                    redrawWindow(WIN,m,count1,count2,b)
                    time.sleep(1)
                    match=m.guess(r1,c1,r2,c2)
                    if match==True:
                        b[spot1].found=True
                        b[spot2].found=True
                    else:
                        b[spot1].text=''
                        b[spot2].text=''
                over=m.checkGameOver(b)
                if over==True:
                    gameOver(m,b,"YOU WON!")
                    count2+=1
                    count1+=1
                else:
                    if m.lifes<0:
                        over=True
                        gameOver(m,b,"BETTER LUCK NEXT TIME")
                        count2+=1
            while over and run:
                for event in pygame.event.get():
                    if event.type==pygame.QUIT:
                        run=False
                        pygame.quit() 
                    if event.type==pygame.MOUSEBUTTONDOWN:
                        pos=pygame.mouse.get_pos()
                        if b[-1].isOverSquare(pos):
                            over=False
                            run=False
                            pygame.quit() 
                        if b[-2].isOverSquare(pos):
                            over=False
                            restart=True
                            b=b[:len(b)-2]

                
    pygame.quit()
   

if __name__ =="__main__":
    main()
        
            
```
