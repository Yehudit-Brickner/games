```python
import random
import time
import pygame

class button():
    def __init__(self,color,x,y,width,height,tsize,text=''):
        self.color=color
        self.x=x
        self.y=y
        self.width=width
        self.height=height
        self.text=text
        self.tsize=tsize
        

    def drawbutton(self,WIN,outline=None):
        if outline:
            pygame.draw.rect(WIN,outline,(self.x-2,self.y-2,self.width+4,self.height+4),0)
        pygame.draw.rect(WIN,self.color,(self.x,self.y,self.width,self.height),0)
        if self.text != '':
            myfont = pygame.font.SysFont('Comic Sans MS', self.tsize)
            text = myfont.render(str(self.text), 1, (0, 0, 0))
            WIN.blit(text,(self.x+(self.width/2-text.get_width()/2),self.y+(self.height/2-text.get_height()/2)))

    def isOverSquare(self,pos):
        if pos[0]>self.x and pos[0]<self.x+self.width:
            if pos[1]>self.y and pos[1]<self.y+self.height:
                return True
        return False




class mastermind():
    def __init__(self):
        self.color=["pink=0","red=1","orange=2","yellow=3","green=4","blue=5","purple=6","brown=7","black=8","white=9"]
        self.colors=[PINK,RED,ORANGE,YELLOW,GREEN,BLUE,PURPLE,BROWN,BLACK,WHITE]
        self.gameover=False
        self.win=False
        self.count=0
        self.code=[]
        self.board=[]
        self.createBoard()
        self.createCode()
        self.clicked=0
        self.gamesplayed=0
        self.gameswon=0
    
    def createBoard(self):
        t=[0]*8
        for i in range(8):
            self.board.append(t)

    def createCode(self):
        for i in range(4):
            self.code.append(self.colors[random.randint(0,len(self.colors)-1)])

    def printCode(self):
        s=""
        for i in range(len(self.code)):
            s+=self.code[i]
            s+="  "
        print(s)

    def guess(self,temp):
        
        for i in range(4):
            if temp[i]==self.code[i]:
                temp.append(RED)
            else:
                ch=0
                for j in range(4):
                    if i!=j:
                        if temp[i]==self.code[j]:
                            ch=YELLOW
                temp.append(ch)
        self.board[self.count]=temp
        self.count+=1
        self.checkGameOver()
        temp=temp[0:4]
        
    def printBoard(self):
        for i in range(len(self.board)):
            s=""
            for j in range(len(self.board[i])):
                s+=self.board[i][j]
                s+="  "
            print(s)
        print()

    def checkGameOver(self):
        row=self.count-1
        c=0
        for i in range(4,8):
            if self.board[row][i]==RED:
                c+=1
        if c==4:
            self.gameover=True
            self.win=True
        else:
            if self.count==8:
                self.gameover=True

    def newGame(self):
        self.gameover=False
        self.win=False
        self.count=0
        self.code=[]
        self.board=[]
        self.clicked=0
        self.createBoard()
        self.createCode()


pygame.init()
pygame.font.init()
WIDTH,HEIGHT =500,650  
WIN=pygame.display.set_mode((WIDTH,HEIGHT))
pygame.display.set_caption("MASTERMIND")
PINK=(255,0,127)
RED=(255,0,0)
ORANGE=(255,185,0)
YELLOW=(255,255,0)
GREEN=(0,205,0)
BLUE=(0,0,255)
PURPLE=(120,0,160)
BROWN=(102,51,0)
WHITE=(255,255,255)
BLACK=(0,0,0)
LIGHTGRAY=(200,200,200)
LIGHTBLUE=(50,150,255)
DARKBLUE=(0,0,102)

colors=[PINK,RED,ORANGE,YELLOW,GREEN,BLUE,PURPLE,BROWN,BLACK,WHITE]



def redrawWindow(w,g,guess,buttonColors):
    w.fill(LIGHTGRAY)
    myfont = pygame.font.SysFont('Comic Sans MS', 40)
    text=myfont.render("MasterMind",1,BLACK )
    WIN.blit(text,(150 +(100-text.get_width()/2),15+(15-text.get_height()/2)))
    myfont = pygame.font.SysFont('Comic Sans MS', 20)
    text=myfont.render("GUESS:",1,BLACK )
    WIN.blit(text,(50 +(50-text.get_width()/2),460+(15-text.get_height()/2)))

    txt="games won "
    txt+=str(g.gameswon)
    txt+=" / "
    txt+=str(g.gamesplayed)
    text=myfont.render(txt,1,BLACK )
    WIN.blit(text,(200 +(50-text.get_width()/2),600+(15-text.get_height()/2)))


    x=150
    y=460
    for i in range(4):
        pygame.draw.rect(w,LIGHTBLUE,(x,y,50,30),2)  
        x+=60

    y=80
    for i in range(8):
        x=120
        for j in range(4):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),20)  
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),20)
            x+=50
        for j in range(4,8):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),10)
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),10)
            x+=30
        y+=50
    
    x=152
    y=462
    for i in range(len(guess)):
        if guess[i]!=0:
            pygame.draw.rect(w,guess[i],(x,y,46,26))  
        x+=60
    
    for i in buttonColors:
        i.drawbutton(w)
    
    pygame.display.update()


def gameOver(w,g,guess,buttonColors,t):
    w.fill(LIGHTGRAY)

    myfont = pygame.font.SysFont('Comic Sans MS', 40)
    text=myfont.render("MasterMind",1,BLACK )
    WIN.blit(text,(150 +(100-text.get_width()/2),15+(15-text.get_height()/2)))
    myfont = pygame.font.SysFont('Comic Sans MS', 20)
    text=myfont.render("GUESS:",1,BLACK )
    WIN.blit(text,(50 +(50-text.get_width()/2),460+(15-text.get_height()/2)))

    txt="games won "
    txt+=str(g.gameswon)
    txt+=" / "
    txt+=str(g.gamesplayed)
    text=myfont.render(txt,1,BLACK )
    WIN.blit(text,(200 +(50-text.get_width()/2),600+(15-text.get_height()/2)))

    x=150
    y=460
    for i in range(4):
        pygame.draw.rect(w,LIGHTBLUE,(x,y,50,30),2)  
        x+=60

    y=80
    for i in range(8):
        x=120
        for j in range(4):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),20)  
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),20)
            x+=50
        for j in range(4,8):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),10)
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),10)
            x+=30
        y+=50
    
    x=152
    y=462
    for i in range(len(guess)):
        pygame.draw.rect(w,guess[i],(x,y,46,26))  
        x+=60
    
    for i in buttonColors:
        i.drawbutton(w)
    

    pygame.draw.rect(w,WHITE,(100,200,300,200)) 
    myfont = pygame.font.SysFont('Comic Sans MS', 40)
    text=myfont.render(t,1,BLACK )
    WIN.blit(text,(150 +(100-text.get_width()/2),290+(15-text.get_height()/2)))
    pygame.display.update()
    time.sleep(2)


def playAgain(w,g,guess,buttonColors):

    w.fill(LIGHTGRAY)
    myfont = pygame.font.SysFont('Comic Sans MS', 40)
    text=myfont.render("MasterMind",1,BLACK )
    WIN.blit(text,(150 +(100-text.get_width()/2),15+(15-text.get_height()/2)))
    myfont = pygame.font.SysFont('Comic Sans MS', 20)
    text=myfont.render("GUESS:",1,BLACK )
    WIN.blit(text,(50 +(50-text.get_width()/2),460+(15-text.get_height()/2)))

    txt="games won: "
    txt+=str(g.gameswon)
    txt+=" / "
    txt+=str(g.gamesplayed)
    text=myfont.render(txt,1,BLACK )
    WIN.blit(text,(200 +(50-text.get_width()/2),600+(15-text.get_height()/2)))

    x=150
    y=460
    for i in range(4):
        pygame.draw.rect(w,LIGHTBLUE,(x,y,50,30),2)  
        x+=60

    y=80
    for i in range(8):
        x=120
        for j in range(4):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),20)  
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),20)
            x+=50
        for j in range(4,8):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),10)
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),10)
            x+=30
        y+=50
    
    x=152
    y=462
    for i in range(len(guess)):
        pygame.draw.rect(w,guess[i],(x,y,46,26))  
        x+=60
    
    for i in buttonColors:
        i.drawbutton(w)

    pygame.draw.rect(w,WHITE,(75,200,350,200)) 
    myfont = pygame.font.SysFont('Comic Sans MS', 25)
    text=myfont.render("Do you want to play again?",1,BLACK)
    WIN.blit(text,(175 +(75-text.get_width()/2),250+(15-text.get_height()/2)))
    yes=button(GREEN,150,300,75,75,30,"YES")
    yes.drawbutton(w)
    no=button(RED,300,300,75,75,30,"NO")
    no.drawbutton(w)

    pygame.display.update()
    play=True
    
    while play:
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                play=False
                pygame.quit() 
                return False
            # if event.type==pygame.MOUSEMOTION:
            #     pos=pygame.mouse.get_pos()
            #     if yes.isOverSquare(pos):
            #         yes.color=(0,155,155)
            #     else:
            #         yes.color=GREEN
            #     if no.isOverSquare(pos):
            #         no.color=(155,0,0)
            #     else:
            #         no.color=RED
            if event.type==pygame.MOUSEBUTTONDOWN:
                pos=pygame.mouse.get_pos()
                if yes.isOverSquare(pos):
                    return True
                if no.isOverSquare(pos):
                    return False
            pygame.display.update()
                

def singleguess(temp,col):
    added=False
    for i in range(len(temp)):
        if added==False:
            if temp[i]==0:
                temp[i]=col
                added=True
    if added==False:
        temp.append(col)



def overGuess(pos,temp,m):
    if len(temp)>=1:
        if 150<=pos[0]<=200 and 460<=pos[1]<=490:
            if temp[0]!=0:
                temp[0]=0
                m.clicked-=1
    if len(temp)>=2:
        if 210<=pos[0]<=260 and 460<=pos[1]<=490:
            if temp[1]!=0:
                temp[1]=0
                m.clicked-=1
    if len(temp)>=3:
        if 270<=pos[0]<=320 and 460<=pos[1]<=490:
            if temp[2]!=0:
                temp[2]=0
                m.clicked-=1
    if len(temp)==4:
        if 330<=pos[0]<=380 and 460<=pos[1]<=490:
            if temp[3]!=0:
                temp[3]=0
                m.clicked-=1


def guessButton(w,g,guess,buttonColors):
    w.fill(LIGHTGRAY)
    myfont = pygame.font.SysFont('Comic Sans MS', 40)
    text=myfont.render("MasterMind",1,BLACK )
    WIN.blit(text,(150 +(100-text.get_width()/2),15+(15-text.get_height()/2)))
    myfont = pygame.font.SysFont('Comic Sans MS', 20)
    text=myfont.render("GUESS:",1,BLACK )
    WIN.blit(text,(50 +(50-text.get_width()/2),460+(15-text.get_height()/2)))

    txt="games won "
    txt+=str(g.gameswon)
    txt+=" / "
    txt+=str(g.gamesplayed)
    text=myfont.render(txt,1,BLACK )
    WIN.blit(text,(200 +(50-text.get_width()/2),600+(15-text.get_height()/2)))


    x=150
    y=460
    for i in range(4):
        pygame.draw.rect(w,LIGHTBLUE,(x,y,50,30),2)  
        x+=60

    y=80
    for i in range(8):
        x=120
        for j in range(4):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),20)  
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),20)
            x+=50
        for j in range(4,8):
            if g.board[i][j]!=0:
                pygame.draw.circle(w,g.board[i][j],(x,y),10)
            else:
                pygame.draw.circle(w,LIGHTBLUE,(x,y),10)
            x+=30
        y+=50
    
    x=152
    y=462
    for i in range(len(guess)):
        if guess[i]!=0:
            pygame.draw.rect(w,guess[i],(x,y,46,26))  
        x+=60
    
    for i in buttonColors:
        i.drawbutton(w)

    gButton=button(WHITE,400,460,50,30,20,"guess")
    gButton.drawbutton(w)
    
    pygame.display.update()

    play=True
    while play:
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                play=False
                pygame.quit() 
                return False
            if event.type==pygame.MOUSEBUTTONDOWN:
                mousePressed=pygame.mouse.get_pressed()
                if mousePressed[0]:
                    pos=pygame.mouse.get_pos()
                    if gButton.isOverSquare(pos):
                        # check guess
                        g.guess(guess)
                        play=False
                if mousePressed[2]:
                    pos=pygame.mouse.get_pos()
                    overGuess(pos,guess,g)
                    if g.clicked<4:
                        play=False 






def main():
    buttonColors=[]
    y=520
    x=100
    count=0
    for i in range(len(colors)):
        count+=1
        buttonColors.append(button(colors[i],x,y,50,30,0))
        x+=60
        if count==5:
            y=570
            x=100

    m=mastermind()
    run=True
    temp=[]
    click=False
    while(run):
        redrawWindow(WIN,m,temp,buttonColors)
        if m.gameover==True:
            if m.win==True:
                m.gamesplayed+=1
                m.gameswon+=1
                gameOver(WIN,m,temp,buttonColors,"YOU WON!")
                m.newGame()
                run=playAgain(WIN,m,temp,buttonColors)
            else:
                m.gamesplayed+=1
                gameOver(WIN,m,temp,buttonColors,"you lost :(")
                m.newGame() 
                run=playAgain(WIN,m,temp,buttonColors)
        m.clicked=len(temp)
        if run==False:
            m.clicked=5
        else:
            redrawWindow(WIN,m,temp,buttonColors)
        while m.clicked<4:     
            for event in pygame.event.get():
                if event.type==pygame.QUIT:
                    run=False
                    pygame.quit() 
                    m.clicked=5   
                if event.type==pygame.MOUSEBUTTONDOWN:
                    mousePressed=pygame.mouse.get_pressed()
                    if mousePressed[0]:
                        pos=pygame.mouse.get_pos()
                        if m.clicked<4:
                            for i in range(len(buttonColors)):
                                if buttonColors[i].isOverSquare(pos):
                                    # temp.append(m.colors[i])
                                    singleguess(temp,m.colors[i])
                                    m.clicked+=1
                        if m.clicked==4:
                            conti=guessButton(WIN,m,temp,buttonColors)
                            if conti==False:
                                run=False
                                pygame.quit() 
                                m.clicked=5
                            elif m.clicked==4:
                                temp=[]
                            if run:
                                redrawWindow(WIN,m,temp,buttonColors)
                        if run:    
                            redrawWindow(WIN,m,temp,buttonColors)
                    elif mousePressed[2]:
                        pos=pygame.mouse.get_pos()
                        overGuess(pos,temp,m)
                        redrawWindow(WIN,m,temp,buttonColors)

                    
    pygame.quit() 



if __name__ =="__main__":
    main()

```
