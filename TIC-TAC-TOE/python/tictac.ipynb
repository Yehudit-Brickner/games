```python
import pygame
import time

from turtle import color

pygame.init()
pygame.font.init()
WIDTH,HIGHT =500,600  
WIN=pygame.display.set_mode((WIDTH,HIGHT))
pygame.display.set_caption("TIC-TAC-TOE")

class button():
    def __init__(self,color,x,y,width,height,name,tsize,text=''):
        self.color=color
        self.x=x
        self.y=y
        self.width=width
        self.height=height
        self.text=text
        self.name=name
        self.tsize=tsize

    def draw(self,WIN,outline=None):
        if outline:
            pygame.draw.rect(WIN,outline,(self.x-2,self.y-2,self.width+4,self.height+4),0)
        pygame.draw.rect(WIN,self.color,(self.x,self.y,self.width,self.height),0)
        if self.text != '':
            myfont = pygame.font.SysFont('Comic Sans MS', self.tsize)
            text = myfont.render(self.text, 1, (0, 0, 0))
            WIN.blit(text,(self.x+(self.width/2-text.get_width()/2),self.y+(self.height/2-text.get_height()/2)))

    def isOver(self,pos):
        if pos[0]>self.x and pos[0]<self.x+self.width:
            if pos[1]>self.y and pos[1]<self.y+self.height:
                return True
        return False

    def redrawWindow(buttons):
        WIN.fill((51,153,255))
        for i in range(len(buttons)):
            buttons[i].draw(WIN,(0,0,0))
       



def checkboard(b, color1, color2):
    # print("checking board")
    for i in range(3):
        if b[0][i].color==b[1][i].color and b[0][i].color==b[2][i].color:
            if b[0][i].color==color1.color or b[0][i].color==color2.color:
                return True
    for i in range(3):
        if b[i][0].color==b[i][1].color and b[i][0].color==b[i][2].color:
            if b[i][0].color==color1.color or b[i][0].color==color2.color:
                return True
    if b[0][0].color==b[1][1].color and b[0][0].color==b[2][2].color:
        if b[0][0].color==color1.color or b[0][0].color==color2.color:
            return True
    if b[2][0].color==b[1][1].color and b[2][0].color==b[0][2].color:
        if b[2][0].color==color1.color or b[2][0].color==color2.color:
            return True 
    return False

def findwinner(b,color1,color2):
    # print("looking for winner")
    # checking rows
    for i in range(3):
        if b[0][i].color==b[1][i].color and b[0][i].color==b[2][i].color:
            if b[0][i].color==color1.color:
                # print("the winner is", color1.name)
                return color1
            elif b[0][i].color==color2.color:
                # print("the winner is", color2.name)
                return color2
    # checking cols
    for i in range(3):
        if b[i][0].color==b[i][1].color and b[i][0].color==b[i][2].color:
            if b[i][0].color==color1.color:
                # print("the winner is", color1.name)
                return color1
            elif b[i][2].color==color2.color:
                # print("the winner is", color2.name)
                return color2
    if b[0][0].color==b[1][1].color and b[0][0].color==b[2][2].color:
        if b[0][0].color==color1.color:
            # print("the winner is", color1.name)
            return color1 
        elif b[0][0].color==color2.color:
            # print("the winner is", color2.name)
            return b[0][0].color
    if b[2][0].color==b[1][1].color and b[2][0].color==b[0][2].color:
        if b[2][0].color==color1.color:
            # print("the winner is", color1.name)
            return color1
        elif b[2][0].color==color2.color:
            # print("the winner is", color2.name)
            return color2
    return None

def cleargame(b):
    b=b[0:13]
    for i in range(9):
        b[i].color=(255,255,102)
        b[i].text=''
    return b

class players():
    def __init__(self,color,name,ch):
        self.color=color
        self.name=name
        self.ch=ch
        self.score=0

def main():
    button1= button((255,255,102),20,340,140,140,"b1",100)
    button2= button((255,255,102),180,340,140,140,"b2",100)
    button3= button((255,255,102),340,340,140,140,"b3",100)
    button4= button((255,255,102),20,180,140,140,"b4",100)
    button5= button((255,255,102),180,180,140,140,"b5",100)
    button6= button((255,255,102),340,180,140,140,"b6",100)
    button7= button((255,255,102),20,20,140,140,"b7",100)
    button8= button((255,255,102),180,20,140,140,"b8",100)
    button9= button((255,255,102),340,20,140,140,"b9",100)
    p1=button((255,255,255),100,520,70,30,"p1",15,"player1")
    p2=button((255,255,255),400,520,70,30,"p2",15,"player2")
    s1=button((255,255,255),100,550,70,30,"s1",15,"0")
    s2=button((255,255,255),400,550,70,30,"s2",15,"0")
    green=players((0,255,89),"green",'X')
    blue=players((0,85,255),"blue",'O')
    
    end=0
    no=0
    yes=0
    count=0
    winner=0

    gameover=False
    run=True 
    newgame=False

    buttons=[[button7,button8,button9],[button4,button5,button6],[button1,button2,button3]]
    b=[button1,button2,button3,button4,button5,button6,button7,button8,button9,p1,p2,s1,s2]
    
    while run:
        button.redrawWindow(b)
        pygame.display.update()
        if newgame==True:
            count=0
            cleargame(b)
            newgame=False
            button.redrawWindow(b)
            pygame.display.update()
            print("starting new game")
        if count>2:
            gameover=checkboard(buttons,green,blue)
        if gameover==True:
            count=0
            gameover=False
            win=findwinner(buttons,green,blue)
            if win!=None:
                win.score+=1
                if win==green:
                    s1.text=str(green.score)  
                else:
                    s2.text=str(blue.score)  
                ans="the winner is "
                ans+=win.name
                ans+="!"
                print(ans)
                win=None
                winner=button((255,255,255),50,200,400,200,"winner",40,ans)
                b.append(winner)
                button.redrawWindow(b)
                pygame.display.update()
                time.sleep(1)
                b=cleargame(b)
                end=button((255,255,255),20,100,480,100,"end",30,"Do you want to play again?")
                yes=button((0,255,0),200,510,50,50,"yes",30,"yes")
                no=button((255,0,0),280,510,50,50,"no",30,"no")
                b.append(end)
                b.append(yes)
                b.append(no)
                button.redrawWindow(b)
                pygame.display.update()

       
        if count==9:
            count=0
            ans="TIE GAME!"
            print(ans)
            tie=button((255,255,255),50,200,400,200,"tie",40,ans)
            b.append(tie)
            button.redrawWindow(b)
            pygame.display.update()
            time.sleep(1)
            b=cleargame(b)
            end=button((255,255,255),20,100,480,100,"end",30,"Do you want to play again?")
            yes=button((0,255,0),200,510,50,50,"yes",30,"yes")
            no=button((255,0,0),280,510,50,50,"no",30,"no")
            b.append(end)
            b.append(yes)
            b.append(no)
            button.redrawWindow(b)
            pygame.display.update()

        if count%2==0:
            player=green
        else:
            player=blue


        for event in pygame.event.get():
            pos=pygame.mouse.get_pos()
            if event.type==pygame.QUIT:
                run=False
                pygame.quit()
            
            if event.type == pygame.MOUSEBUTTONDOWN:
                if button1.isOver(pos):
                    if button1.color!=green.color and button1.color!=blue.color :
                        button1.color=player.color
                        button1.text=player.ch
                        count+=1
                elif button2.isOver(pos):
                    if button2.color!=green.color and button2.color!=blue.color:
                        button2.color=player.color
                        button2.text=player.ch
                        count+=1   
                elif button3.isOver(pos):
                    if button3.color!=green.color and button3.color!=blue.color:
                        button3.color=player.color
                        button3.text=player.ch
                        count+=1
                elif button4.isOver(pos):
                    if button4.color!=green.color and button4.color!=blue.color:
                        button4.color=player.color
                        button4.text=player.ch
                        count+=1
                elif button5.isOver(pos):
                    if button5.color!=green.color and button5!=blue.color:
                        button5.color=player.color
                        button5.text=player.ch
                        count+=1
                elif button6.isOver(pos):
                    if button6.color!=green.color and button6.color!=blue.color:
                        button6.color=player.color
                        button6.text=player.ch
                        count+=1
                elif button7.isOver(pos):
                    if button7.color!=green.color and button7.color!=blue.color :
                        button7.color=player.color
                        button7.text=player.ch
                        count+=1
                elif button8.isOver(pos):
                    if button8.color!=green.color and button8.color!=blue.color :
                        button8.color=player.color
                        button8.text=player.ch
                        count+=1
                elif button9.isOver(pos):
                    if button9.color!=green.color and button9.color!=blue.color:
                        button9.color=player.color
                        button9.text=player.ch
                        count+=1
                if no in b:
                    if no.isOver(pos):
                        run=False
                if yes in b:
                    if yes.isOver(pos):
                        b=cleargame(b)
                        count=0
                        newgame=True
                

            if event.type==pygame.MOUSEMOTION:

                if button1.color!=green.color and button1.color!=blue.color:
                    if button1.isOver(pos):
                        button1.color=(255,204,204)
                    else:
                        button1.color=(255,255,102)
                if button2.color!=green.color and button2.color!=blue.color:
                    if button2.isOver(pos):
                        button2.color=(255,204,204)
                    else:
                        button2.color=(255,255,102)
                if button3.color!=green.color and button3.color!=blue.color:
                    if button3.isOver(pos):
                        button3.color=(255,204,204)
                    else:
                        button3.color=(255,255,102)
                if button4.color!=green.color and button4.color!=blue.color:
                    if button4.isOver(pos):
                        button4.color=(255,204,204)
                    else:
                        button4.color=(255,255,102)
                if button5.color!=green.color and button5.color!=blue.color :
                    if button5.isOver(pos):
                        button5.color=(255,204,204)
                    else:
                        button5.color=(255,255,102)
                if button6.color!=green.color and button6.color!=blue.color:
                    if button6.isOver(pos):
                        button6.color=(255,204,204)
                    else:
                        button6.color=(255,255,102)
                if button7.color!=green.color and button7.color!=blue.color:
                    if button7.isOver(pos):
                        button7.color=(255,204,204)
                    else:
                        button7.color=(255,255,102)
                if button8.color!=green.color and button8.color!=blue.color:
                    if button8.isOver(pos):
                        button8.color=(255,204,204)
                    else:
                        button8.color=(255,255,102)
                if button9.color!=green.color and button9.color!=blue.color:
                    if button9.isOver(pos):
                        button9.color=(255,204,204)
                    else:
                        button9.color=(255,255,102)
                if yes in b:
                    if yes.isOver(pos):
                        yes.color=(153,255,153)
                    else:
                        yes.color=(0,255,0)
                if no in b:
                    if no.isOver(pos):
                        no.color=(255,153,153)
                    else:
                        no.color=(255,0,0)
    # print("exited while loop")
    if(count==9):
        print("tie game")
    pygame.quit()            
          

if __name__ =="__main__":
    main()


```
