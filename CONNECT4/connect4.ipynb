```python

# connect4 game1
import pygame
import time
from turtle import color

pygame.init()
pygame.font.init()
WIDTH,HIGHT =500,550  
WIN=pygame.display.set_mode((WIDTH,HIGHT))
pygame.display.set_caption("CONNECT 4")

class button():
    def __init__(self,color,x,y,width,height,tsize,text=''):
        self.color=color
        self.x=x
        self.y=y
        self.width=width
        self.height=height
        self.text=text
        self.tsize=tsize

    def drawSquare(self,WIN,outline=None):
        if outline:
            pygame.draw.rect(WIN,outline,(self.x-2,self.y-2,self.width+4,self.height+4),0)
        pygame.draw.rect(WIN,self.color,(self.x,self.y,self.width,self.height),0)
        if self.text != '':
            myfont = pygame.font.SysFont('Comic Sans MS', self.tsize)
            text = myfont.render(self.text, 1, (0, 0, 0))
            WIN.blit(text,(self.x+(self.width/2-text.get_width()/2),self.y+(self.height/2-text.get_height()/2)))

    def drawCircle(self,WIN,outline=None):
        pygame.draw.ellipse(WIN,self.color,(self.x,self.y,self.width,self.height),0)

    def isOverSquare(self,pos):
        if pos[0]>self.x and pos[0]<self.x+self.width:
            if pos[1]>self.y and pos[1]<self.y+self.height:
                return True
        return False
        
    def redrawWindow(s,c):
        WIN.fill((0,0,255))
        for i in range(len(c)):
            for j in range(len(c[i])):
                c[i][j].drawCircle(WIN,(0,0,0))
        for i in range(len(s)):
            s[i].drawSquare(WIN,(0,0,0))
        
    def createButtonsCircle():
        b=[]
        c=[50,110,170,230,290,350,410]
        r=[200,260,320,380,440,500]
        w=(255,255,255)
        for j in range(len(r)):
            row=[]
            for i in range(len(c)):
                row.append(button((w),c[i],r[j],40,40,0))
            b.append(row)
        return b

    def createButtonsSquare():
        b=[]
        c=[50,110,170,230,290,350,410]
        w=(0,255,155)
        for j in range(7):
            txt="column "
            txt+=str(j+1)
            b.append(button((w),c[j],150,50,30,10,txt))
        return b
   
    def checkPreesedButton(bs,bc,num,g1): 
        if g1.column[num-1]<6:
            txt=""
            if g1.turn==g1.player1.turn:
                txt+=g1.player1.name
            else:
                txt+=g1.player2.name
            txt+=" do you want to pick column "
            txt+=str(num)
            bs.append(button((120,120,120),30,150,450,40,20,txt))
            bs.append(button((0,255,0),100,200,50,50,20,"yes"))
            bs.append(button((255,0,0),350,200,50,50,20,"no"))
            button.redrawWindow(s=bs,c=bc)
            pygame.display.update()

class player():
    def __init__(self,name,color,ch,turn):
        self.name=name
        self.color=color
        self.ch=ch
        self.score=0
        self.turn=turn

class c4():
    def __init__(self,cboard):
        self.cboard=cboard
        self.listboard=[]
        for i in range(6):
            r=[]
            for i in range(7):
                r.append(' ')
            self.listboard.append(r)
        self.column=[0]*7
        red=(255,0,0)
        yellow=(255,255,0)
        self.player1=player("red",red,'r',1)
        self.player2=player("yellow",yellow,'y',2)
        self.turn=1
    
    def updateGameBoard(self,col,bs):
        if self.turn==1:
            self.listboard[6-self.column[col]][col]=self.player1.ch
            self.cboard[6-self.column[col]][col].color=self.player1.color
            a=self.checkBoard()
            if a[1]==self.player1.ch or a[1]==self.player2.ch:
                return a[1]
            self.turn=2
            bs[11].text="YELLOW turn"
        else:
            self.listboard[6-self.column[col]][col]=self.player2.ch
            self.cboard[6-self.column[col]][col].color=self.player2.color
            a=self.checkBoard()
            if a[1]==self.player1.ch or a[1]==self.player2.ch:
                return a[1]
            self.turn=1
            bs[11].text="RED turn"
   
    def checkBoard(self):
        #rows
        for i in range(6):
            if self.listboard[i][0]==self.listboard[i][1]==self.listboard[i][2]==self.listboard[i][3]:
                if self.listboard[i][0]!=' ':
                    # print("4 in row")
                    return (True,self.listboard[i][0])
            if self.listboard[i][1]==self.listboard[i][2]==self.listboard[i][3]==self.listboard[i][4]:
                if self.listboard[i][1]!=' ':
                    # print("4 in row")
                    return (True,self.listboard[i][1])
            if self.listboard[i][2]==self.listboard[i][3]==self.listboard[i][4]==self.listboard[i][5]:
                if self.listboard[i][2]!=' ':
                    # print("4 in row")
                    return (True,self.listboard[i][2])
            if self.listboard[i][3]==self.listboard[i][4]==self.listboard[i][5]==self.listboard[i][6]:
                if self.listboard[i][3]!=' ':
                    # print("4 in row")
                    return (True,self.listboard[i][3])
        #cols
        for i in range(7):
            if self.listboard[0][i]==self.listboard[1][i]==self.listboard[2][i]==self.listboard[3][i]:            
                if self.listboard[0][i]!=' ':
                    # print("4 in col")
                    return (True,self.listboard[0][i])
            if self.listboard[1][i]==self.listboard[2][i]==self.listboard[3][i]==self.listboard[4][i]:            
                if self.listboard[1][i]!=' ':
                    # print("4 in col")
                    return (True,self.listboard[1][i])
            if self.listboard[2][i]==self.listboard[3][i]==self.listboard[4][i]==self.listboard[5][i]:            
                if self.listboard[2][i]!=' ':
                    # print("4 in col")
                    return (True,self.listboard[2][i])
            
        #diags:
        #short1
        if self.listboard[3][0]==self.listboard[2][1]==self.listboard[1][2]==self.listboard[0][3]:            
                if self.listboard[3][0]!=' ':
                    # print("4 in diag short1")
                    return (True,self.listboard[3][0]) 
        #short2
        if self.listboard[5][3]==self.listboard[4][4]==self.listboard[3][5]==self.listboard[2][6]:            
                if self.listboard[5][3]!=' ':
                    # print("4 in diag short2")
                    return (True,self.listboard[5][3]) 
        #short3
        if self.listboard[2][0]==self.listboard[3][1]==self.listboard[4][2]==self.listboard[5][3]:            
                if self.listboard[2][0]!=' ':
                    # print("4 in diag short 3")
                    return (True,self.listboard[2][0]) 
        #short4
        if self.listboard[0][3]==self.listboard[1][4]==self.listboard[2][5]==self.listboard[3][6]:            
                if self.listboard[0][3]!=' ':
                    # print("4 in diag short 4")
                    return (True,self.listboard[0][3])
        #medium1
        for i in range(2):
            if self.listboard[4-i][i]==self.listboard[3-i][1+i]==self.listboard[2-i][2+i]==self.listboard[1-i][3+i]:            
                if self.listboard[4-i][i]!=' ':
                    # print("4 in diag medium1")
                    return (True,self.listboard[4-1][0+i]) 
        #medium2
        for i in range(2):
            if self.listboard[5-i][2+i]==self.listboard[4-i][3+i]==self.listboard[3-i][4+i]==self.listboard[2-i][5+i]:            
                if self.listboard[5-i][2+i]!=' ':
                    # print("4 in diag medium2")
                    return (True,self.listboard[5-i][2+i]) 

        #medium3
        for i in range(2):
            if self.listboard[1+i][i]==self.listboard[2+i][1+i]==self.listboard[3+i][2+i]==self.listboard[4+i][3+i]:            
                if self.listboard[1+i][i]!=' ':
                    # print("4 in diag medium3")
                    return (True,self.listboard[1+i][i]) 
        #medium4
        for i in range(2):
            if self.listboard[i][2+i]==self.listboard[1+i][3+i]==self.listboard[2+i][4+i]==self.listboard[3+i][5+i]:            
                if self.listboard[i][2+i]!=' ':
                    # print("4 in diag medium4")
                    return (True,self.listboard[1+i][i])  
        #long1
        for i in range(3):
             if self.listboard[5-i][i]==self.listboard[4-i][i+1]==self.listboard[3-i][2+i]==self.listboard[2-i][i+3]:            
                if self.listboard[5-i][i]!=' ':
                    # print("4 in diag long1")
                    return (True,self.listboard[5-i][i])
        
        #long2
        for i in range(3):
             if self.listboard[5-i][1+i]==self.listboard[4-i][2+1]==self.listboard[3-i][3+i]==self.listboard[2-i][4+i]:            
                if self.listboard[5-i][1+i]!=' ':
                    # print("4 in diag long2")
                    return (True,self.listboard[5-i][1+i])
        
        #long3
        for i in range(3):
            if self.listboard[i][i]==self.listboard[1+i][1+i]==self.listboard[2+i][2+i]==self.listboard[3+i][3+i]:            
                if self.listboard[i][i]!=' ':
                    # print("4 in diag long3")
                    return (True,self.listboard[i][i])
        #long4
        for i in range(3):
            if self.listboard[i][1+i]==self.listboard[1+i][2+i]==self.listboard[2+i][3+i]==self.listboard[3+i][4+i]:            
                if self.listboard[i][1+i]!=' ':
                    # print("4 in diag long4")
                    return (True,self.listboard[i][1+i])
        return (False,'0')

    def clearGame(self,bs):
        for i in range(6):
            for j in range(7):
                self.cboard[i][j].color=(255,255,255)
                self.listboard[i][j]=' '
        self.column=[0]*7
        for i in range(7):
            bs[i].color=(0,255,0) 


def main():

    bc=button.createButtonsCircle()
    bs=button.createButtonsSquare()
    bs.append(button((255,0,0),100,20,70,30,20,"red"))
    bs.append(button((255,255,0),350,20,70,30,20,"yellow"))
    bs.append(button((255,0,0),100,50,70,30,20,"0"))
    bs.append(button((255,255,0),350,50,70,30,20,"0"))
    bs.append(button((50,50,50),190,75,140,50,15,"RED turn"))
    run=True 
    g1=c4(bc)
    clicked=-1
    won=False
    yesno=False
    gray=(50,50,50)
    tie=False
    while run:
        if (won==True or tie==True) and yesno==True:
            # print("clearing game")
            g1.clearGame(bs)
            clicked=None
            won=False
            tie=False
            yesno=False
        if won==False and tie==False:
            count=0
            for i in range(7):
                if bs[i].color==gray:
                    count+=1
            if count==7:
                tie=True
                # print(tie)
                t=button((255,255,255),50,200,400,200,40,"TIE GAME")
                bs.append(t)
                button.redrawWindow(s=bs,c=bc)
                pygame.display.update()
                time.sleep(1)
                bs=bs[0:11]
                bs.append(button((120,120,120),30,200,450,40,20,"Do you want to play again?"))
                bs.append(button((0,255,0),100,250,50,50,20,"yes"))
                bs.append(button((255,0,0),350,250,50,50,20,"no"))
                button.redrawWindow(s=bs,c=bc)
                pygame.display.update()

        button.redrawWindow(s=bs,c=bc)
        pygame.display.update()
        for event in pygame.event.get():
            pos=pygame.mouse.get_pos()
            if event.type==pygame.QUIT:
                run=False
                pygame.quit()

            if event.type == pygame.MOUSEBUTTONDOWN:
                
                if bs[0].isOverSquare(pos):
                    clicked=0
                    button.checkPreesedButton(bs,bc,1,g1)
                    
                if bs[1].isOverSquare(pos):
                    clicked=1
                    button.checkPreesedButton(bs,bc,2,g1)

                if bs[2].isOverSquare(pos):
                    clicked=2
                    button.checkPreesedButton(bs,bc,3,g1)

                if bs[3].isOverSquare(pos):
                    clicked=3
                    button.checkPreesedButton(bs,bc,4,g1)

                if bs[4].isOverSquare(pos):
                    clicked=4
                    button.checkPreesedButton(bs,bc,5,g1)

                if bs[5].isOverSquare(pos):
                    clicked=5
                    button.checkPreesedButton(bs,bc,6,g1)

                if bs[6].isOverSquare(pos):
                    clicked=6
                    button.checkPreesedButton(bs,bc,7,g1)

                
                if len(bs)>14:
                    if bs[13].isOverSquare(pos):
                        #yes for picking spot
                        if won==False and tie==False:
                            bs=bs[0:12]
                            if clicked!=None:
                                g1.column[clicked]+=1
                                if g1.column[clicked]==6:
                                    bs[clicked].color=(50,50,50)
                                ans = g1.updateGameBoard(clicked,bs)
                                if ans!=None:
                                    won=True
                                    txt="The winner is "
                                    if ans=='r':
                                        txt+="red"
                                        g1.player1.score+=1
                                        bs[9].text=str(g1.player1.score)
                                    else:
                                        txt +="yellow"
                                        g1.player2.score+=1
                                        bs[10].text=str(g1.player2.score)
                                    winner=button((255,255,255),50,200,400,200,40,txt)
                                    bs.append(winner)
                                    button.redrawWindow(s=bs,c=bc)
                                    pygame.display.update()
                                    time.sleep(1)
                                    bs=bs[0:12]
                                    bs.append(button((120,120,120),30,200,450,40,20,"Do you want to play again?"))
                                    bs.append(button((0,255,0),100,250,50,50,20,"yes"))
                                    bs.append(button((255,0,0),350,250,50,50,20,"no"))
                                    button.redrawWindow(s=bs,c=bc)
                                    pygame.display.update()
                        #yes for playing again
                        else:     
                            if won==True or tie==True:
                                yesno=True
                                bs=bs[0:12]

                    #no
                    elif bs[14].isOverSquare(pos):
                        if won==True or tie==True:
                            yesno=True
                            run=False
                        bs=bs[0:12]
                

                    


            if event.type==pygame.MOUSEMOTION:
                if bs[0].color != gray:
                    if bs[0].isOverSquare(pos):
                        bs[0].color=(150,0,150)
                    else:
                        bs[0].color=(0,255,155)
                if bs[1].color != gray:
                    if bs[1].isOverSquare(pos):
                        bs[1].color=(150,0,150)
                    else:
                        bs[1].color=(0,255,155)
                if bs[2].color != gray:
                    if bs[2].isOverSquare(pos):
                        bs[2].color=(150,0,150)
                    else:
                        bs[2].color=(0,255,155)
                if bs[3].color != gray:
                    if bs[3].isOverSquare(pos):
                        bs[3].color=(150,0,150)
                    else:
                        bs[3].color=(0,255,155)
                if bs[4].color != gray:
                    if bs[4].isOverSquare(pos):
                        bs[4].color=(150,0,150)
                    else:
                        bs[4].color=(0,255,155)
                if bs[5].color != gray:
                    if bs[5].isOverSquare(pos):
                        bs[5].color=(150,0,150)
                    else:
                        bs[5].color=(0,255,155)
                if bs[6].color != gray:
                    if bs[6].isOverSquare(pos):
                        bs[6].color=(150,0,150)
                    else:
                        bs[6].color=(0,255,155)

                if len(bs)>14:
                    if bs[13].isOverSquare(pos):
                        bs[13].color=(153,255,153) 
                    else:
                        bs[13].color=(0,255,0) 
                    
                    if bs[14].isOverSquare(pos):
                        bs[14].color=(255,204,229)
                    else:
                        bs[14].color=(255,0,0)
            
                
    pygame.quit()            



if __name__ =="__main__":
    main()

```
